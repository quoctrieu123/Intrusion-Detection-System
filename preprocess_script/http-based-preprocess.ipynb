```python
links = [
    r"IP-files\HTTP-based\BYPASS-MHDDoS.csv",
    r"IP-files\HTTP-based\CFB-MHDDoS.csv",
    r"IP-files\HTTP-based\DGB-MHDDoS.csv",
    r"IP-files\HTTP-based\HEAD-MHDDoS.csv",
    r"IP-files\HTTP-based\HTTP-Flood-Xerox.csv",
    r"IP-files\HTTP-based\KILLER-MHDDoS.csv",
    r"IP-files\HTTP-based\Mixed-Get-Post-Head.csv",
    r"IP-files\HTTP-based\NULL-MHDDoS.csv",
    r"IP-files\HTTP-based\POST-Flood-MHDDoS.csv",
    r"IP-files\HTTP-based\STOMP-MHDDoS.csv",
]
```

```python
for link in links:
    data = pd.read_csv(link)
    fill_0 = [col for col in data.columns if ("skewness" in col.lower() or "cov" in col.lower()) and data[col].isnull().sum() >0]
    data[fill_0] = data[fill_0].fillna(0)
    cols_has_null = [col for col in data.columns if data[col].isnull().sum() >0]
    if len(cols_has_null) > 0:
        print(f"Warning: {link} still has null values in columns: {cols_has_null}")
    else:
        data.to_csv(link.replace(r"IP-files\HTTP-based", r"ip-files-fillna-lstm").replace(".csv", "-fillna.csv"), index=False)
        print(f"{link} has been processed and saved without null values.")
```
